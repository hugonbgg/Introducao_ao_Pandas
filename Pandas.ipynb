## Alterando os dados

```python
df.head()
```

```python
#criando uma coluna Total do jeito difícil
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```

```python
df.head()
```

```python
#Listando o com maior total
df.sort_values(['Total'], ascending = [False])
```

```python
#removendo a coluna Total
df = df.drop(columns=['Total'])
```

```python
#criando uma coluna Total do jeito fácil
#df.iloc[todas linhas, colunas 4 a 10 não inclusa]
df['Total'] = df.iloc[:,4:10].sum(axis = 1)
```

```python
df.head()
```

```python
#Alterando/reordenando o local da coluna
cols = list(df.columns.values)
```

```python
df = df[cols[0:10] + [cols[-1]] + cols[10:12]]
```

```python
df
```

```python
#salvando o arquivo alterado
df.to_csv('pokemon_data_modificado.csv', index = False, sep = ';')
```

## Filtrando os dados

```python
#filtrando os pokemons do tipo 1 Grass e tipo 2 Poison
df.loc[(df['Type 1'] == 'Grass' ) & (df['Type 2'] == 'Poison')]
```

```python
#filtrando os pokemons do tipo 1 Fire e tipo 2 flying
df.loc[(df['Type 1'] == 'Fire' ) & (df['Type 2'] == 'Flying')]
```

```python
#filtrando e salvando em um dataframe os pokemons do tipo 1 Grass e tipo 2 Poison e HP maior que 70
new_df = df.loc[(df['Type 1'] == 'Grass' ) & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
new_df
```

```python
#Resetando o index do new_df (Se reparar ele manteve o index do df original)
new_df = new_df.reset_index()
new_df
```

```python
#Repare que ele manteve o index anterior como uma coluna chamada index, para remover e naõ criar a coluna, use:
new_df.reset_index(drop = True, inplace = True)
new_df
```

```python
#Filtrando os nomes que possuem mega
df.loc[df['Name'].str.contains('Mega')].head()
```

### Usando Regex

```python
#Usando Regex
#Filtrando os pokemons do tipo 1 que são fire ou grass, o case sensitive foi desativado usando oo re.I
import re

df.loc[df['Name'].str.contains('fire|grass', flags = re.I, regex = True)].head()
df.loc[df['Name'].str.contains('pi[a-z]*', flags = re.I, regex = True)].head()
```

```python
#Pokemons que comecem com Pi
df.loc[df['Name'].str.contains('^pi[a-z]*', flags = re.I, regex = True)].head()
```

```python
#Alterando os valores das colunas a partir de condição
'''Seleciona a coluna que vai ser filtrada, nesse caso a 'Type 1' e o valor usado para filtrar, que vai ser 'Fire', 
depois a que vai ser alterada que vai ser a 'Type 1', e por fim para qual era será alterada, que vai ser 'flamer'
'''

#df = pd.read_csv("pokemon_data_modificado.csv", sep = ';')

df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df
```

```python
#Filtar 'Total' maior que 500 e setar 'Test 1' e 'Test 2' para as colunas Generation e legendary
df.loc[df['Total'] > 500, ['Generation', 'Legendary']] = ['Test 1', 'Test 2']
df
```

### Estatísticas agregadas (GroupBy)


```python
df = pd.read_csv("pokemon_data_modificado.csv", sep = ';')
df.head()
```

```python
#Agrupando os pokemons por Tipo 1 e exibindo a média dos valores, organizando a partir da defesa em ordem decrescente
df.groupby(['Type 1']).mean().sort_values('Defense', ascending = False).head()
```

```python
#Agrupando os pokemons por Tipo 1 e exibindo a média dos valores, organizando a partir do ataque em ordem decrescente
df.groupby(['Type 1']).mean().sort_values('Attack', ascending = False).head()
```

```python
#Criando uma lista com os atributos abaixo e imprimindo as estatitíscas para cada uma delas agrupadas pelos atributos definidos.
atributos = ['HP', 'Attack','Defense']
for atributo in atributos:
    print('Organizando por: '+ atributo)
    print(df.groupby(['Type 1']).mean().sort_values(atributo, ascending = False).head(5))
    print('\n')
```

```python
#Somando todos os atributos por categoria
df.groupby(['Type 1']).sum()
```

```python
#Contando a quantidade de valores em cada categoria, é bom para identificar os NaNs, repare  que a coluna 'Type 2'
#tem bem menos valores que as outras.
df.groupby(['Type 1']).count().head()
```

```python
#Criando uma coluna com a contagem de linha e contando por ela. Daria para contar por outra coluna tbm
df['count'] = 1
df.groupby(['Type 1']).count()['count']
```

```python
#Agrupando e contando por mais de um valor, nesse caso os type 1 e as subcategorias no type 2.
df.groupby(['Type 1','Type 2']).count()['count']
```
